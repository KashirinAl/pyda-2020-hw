## Задание 1
Дана переменная, в которой хранится слово из латинских букв. Напишите код, который выводит на экран:

среднюю букву, если число букв в слове нечетное;
две средних буквы, если число букв четное.

```python
word = str(input('Введите слово '))
L = int(len(word))
Q = int(L / 2) # Середина слова, округлённая до целого
O = str(word)[Q] # Средняя буква
if L % 2 == 0: # Если длина строки чётная, то в L/2 попала левая из двух средних
    P = word[Q + 1]
    print(O, P)
else:
    print(O)
```

## Задание 2
Напишите программу, которая последовательно 
запрашивает у пользователя числа 
(по одному за раз) и после первого нуля выводит 
сумму всех ранее введенных чисел.

```python
A = 9 # ненулевое значение перед началом цикла
S = 0
while A != 0:
    A = int(input('Введите число '))
    S += A
print('Результат: ', S)
```

## Задание 4
У нас есть список, содержащий информацию о среднедневной 
температуре в Фаренгейтах за произвольный период 
по странам (структура данных в примере). 
Необходимо написать код, 
который рассчитает среднюю температуру 
за период в Цельсиях(!) для каждой страны.


```python
countries_temperature = [
    ['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Germany', [57.2, 55.4, 59, 59, 53.6]],
    ['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]
for i in countries_temperature:
    N = i[0] # Название страны
    S = sum(i[1])
    C = len(i[1])
    Q = (S / C - 32) * 5 / 9
    A = round(Q, 1)
    print(N, ' - ', A, " C ")
```

## Задание 5 (необязательное)
Дан поток логов по количеству просмотренных 
страниц для каждого пользователя. 
Вам необходимо написать алгоритм, 
который считает среднее значение просмотров 
на пользователя. Т. е. надо посчитать отношение 
суммы всех просмотров к количеству уникальных пользователей.

```python
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user4,4',
    '2018-04-22,user4,8',
    '2018-05-03,user4,9',
    '2018-05-11,user4,11',
]

A = sorted(stream)
T = '' # Список уникальных пользователей
Q = 0 # Количество уникальных пользователей
S = 0 # Сумма всех посещений
for i in A: # Проходим по записям
    A1 = i.find("user") + 4 # Пользователи различаются только номером, поэтому user - можно обрезать
    A2 = i.rfind(",")
    O = [i[A1:A2], int(i[A2 + 1:len(i)])] # Преобразуем строку в список по найденным маркерам
    S += O[1] # вне зависимости от уникальности суммируем посещения
    if O[0] not in T: # проверяем на уникальность для подсчёта
        Q += 1
        T += O[0]
print('Срднее количество просмотров на уникального пользователя: ', round((S / Q), 2))
        
        
```

## Задание 6 (необязательное)
Необходимо у пользователя запрашивать 
набор чисел разделенных пробелом. 
В результате должны выводиться числа, 
которые повторяются в вводе более одного раза.

```python
print('Введите числа через пробел: ')
A = input('')
S = ''
i = 0
for O in A:
        i += 1
        if i % 2 == 0 and O !=' ': # Проверка, что числа через пробел
            print ('Нужно вводить числа через пробел')
            break 
        else:
            if O in A[0:i - 1] and O !=' ' and O not in S: # Проверка, что число раньше уже встречалось, что это не пробел, и что число ещё не встречается более двух раз
                S = S + O + ' '
print('Результат: ')
if S == '':
    print('Повторений нет')
else:
    print(S)
```
